# Coding practice: a normalizing flow's density

A normalizing flow turns a simple base density (here a standard normal) into a richer one through an invertible transform. The **change-of-variables** rule says how the density transforms:

$$\log p_X(x) = \log p_Z(z) + \log\left|\frac{dz}{dx}\right|, \qquad z = f(x).$$

Here the flow is a simple scaling $z = a\,x$, so the Jacobian is $|dz/dx| = |a|$. Fill in the one TODO, run the cells, and enter the printed number in the check on the course page.

```python
import math

def log_normal_pdf(z):
    """log density of the standard normal N(0, 1) at z (provided)."""
    return -0.5 * (z**2 + math.log(2 * math.pi))

# The flow is z = a * x, a simple invertible scaling.
a = 2.0
```
