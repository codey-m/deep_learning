```python
import math

def log_normal_pdf(z):
    """Log density of the standard normal N(0, 1)."""
    return -0.5 * (z**2 + math.log(2 * math.pi))
```

```python
def log_density(x, a):
    """Return log p_X(x) for the data-to-base map z=a*x."""
    if a == 0:
        raise ValueError("A flow must be invertible, so a cannot be zero.")
    z = a * x
    base = log_normal_pdf(z)
    # TODO: return base plus the log absolute data-to-base Jacobian.
    raise NotImplementedError("Add log|a| to the base log density.")
```

```python
cases = [(0.0, 2.0), (0.0, -2.0), (0.5, 3.0)]
for x, a in cases:
    print(f"x={x:>3}, a={a:>4}: log p_X={log_density(x, a):.4f}")

# The sign of a changes orientation, not density magnitude at x=0.
assert abs(log_density(0.0, 2.0) - log_density(0.0, -2.0)) < 1e-12

# A coarse integral should be close to one.
dx = 0.002
grid = [(-5.0 + i * dx) for i in range(5001)]
integral = sum(math.exp(log_density(x, 2.0)) for x in grid) * dx
print("numerical integral for a=2:", round(integral, 4))
assert abs(integral - 1.0) < 0.01
```
