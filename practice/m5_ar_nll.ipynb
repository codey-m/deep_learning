# Coding practice: stable and masked autoregressive NLL

An autoregressive sequence NLL sums negative log conditional probabilities. In a padded batch, the reportable per-token mean must exclude padding. Complete both functions and use the tests; the course-page checks use changed batches rather than asking you to transcribe a fixed printout.

```python
import math

conditional_probs = [0.5, 0.25, 0.5]
padded_batch = [
    ([0.8, 0.5, 0.25, 1.0], [1, 1, 1, 0]),
    ([0.6, 0.5, 1.0, 1.0], [1, 1, 0, 0]),
]
```

```python
def sequence_nll(probs):
    """Return the summed sequence NLL without multiplying probabilities."""
    if any(p <= 0 or p > 1 for p in probs):
        raise ValueError("Each observed-token probability must lie in (0, 1].")
    # TODO: sum -log(p) directly. Do not multiply first.
    raise NotImplementedError("Return the stable sum of negative logs.")

def masked_token_nll(batch):
    """Return total NLL divided by the number of valid target tokens."""
    total = 0.0
    valid = 0
    for probs, mask in batch:
        for p, keep in zip(probs, mask):
            # TODO: when keep==1, add -log(p) and increment valid.
            pass
    if valid == 0:
        raise ValueError("The batch has no valid target tokens.")
    return total / valid
```

```python
print("sequence summed NLL =", round(sequence_nll(conditional_probs), 4))
print("batch mean NLL per valid token =", round(masked_token_nll(padded_batch), 4))

# Multiplying many tiny probabilities can underflow; summing logs stays finite.
tiny = [1e-200, 1e-200]
print("stable tiny-probability NLL =", sequence_nll(tiny))
assert math.isfinite(sequence_nll(tiny))
```
