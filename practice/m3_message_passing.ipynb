```python
def aggregate_variant(features, neighbors, node, mode="sum", include_self=False):
    values = [features[nbr] for nbr in neighbors[node]]
    if include_self:
        values.append(features[node])
    if mode == "sum":
        return sum(values)
    if mode == "mean":
        return sum(values) / len(values)
    raise ValueError("mode must be 'sum' or 'mean'")

node = 2
sum_value = aggregate_variant(features, neighbors, node, "sum")
mean_value = aggregate_variant(features, neighbors, node, "mean")
self_loop_value = aggregate_variant(features, neighbors, node, "sum", include_self=True)
assert sum_value == 7
assert abs(mean_value - 7 / 3) < 1e-12
assert self_loop_value == 10
print({"sum": sum_value, "mean": round(mean_value, 3), "sum_with_self_loop": self_loop_value})
```
